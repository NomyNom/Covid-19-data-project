```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv("COVID-19_Vaccinations_in_the_United_States_County.csv")
df.head(1000)
```

```python
temp_df = df.loc[df['Recip_County'].isin(['Harris County'])]
fixed_df = temp_df.loc[df['Recip_State'].isin(['TX'])]
fixed_df.head(1000)
```

```python
# Convert the date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Filter the data by Recip_County, Recip_State, and the date range
temp_df = df.loc[(df['Recip_County'] == 'Harris County') & (df['Recip_State'] == 'TX')]

# Filter the data to only include rows between 12/14/2020 and 12/31/2022
filtered_df = temp_df.loc[(temp_df['Date'] >= '2020-12-14') & (temp_df['Date'] <= '2022-12-31')]

filtered_df.head(1000)
```

## Ethnicity Vaccination rates 

```python
df = pd.read_csv("Download_COVID_Vaccination_Progress_By_Race_and_Ethnicity.csv")
df.head()
```

```python
# Convert the date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y 5:00:00 AM')

# Filter the data by Recip_County, Recip_State, and the date range
# temp_df = df.loc[(df['Recip_County'] == 'Harris County') & (df['Recip_State'] == 'TX')]

# Filter the data to only include rows between 12/14/2020 and 12/31/2022
filtered_df = df.loc[(df['Date'] >= '2020-12-14') & (df['Date'] <= '2022-1-1')]

filtered_df.head()
```

```python
df
```

```python
# Dont need column 'Today'
# 'Date_Str was formated in str form and data WAS formatted normally

filtered_df.drop(columns=['Today', 'Date_Str'])
df.drop(columns=['Today', 'Date_Str'])
```

```python
plt.plot(filtered_df['Date'], filtered_df['Asian'], label='Asian')
plt.plot(filtered_df['Date'], filtered_df['Black'], label='Black or African American')
plt.plot(filtered_df['Date'], filtered_df['White'], label='White')
plt.plot(filtered_df['Date'], filtered_df['Hispanic'], label='Latino')
#plt.plot(filtered_df['Date'], filtered_df['AmericanIndian_AlaskanNative'], label='AmericanIndian_AlaskanNative')

plt.title('Vaccination Rates of Different Racial/Ehtnic Groups in Harris County During 2021-2022')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Percentage Vaccinated')

plt.legend() 
plt.show()
```

```python
plt.figure(figsize=(12, 8))
ax = plt.gca()


ax.plot(filtered_df['Date'], filtered_df['Asian'], label='Asian')
ax.plot(filtered_df['Date'], filtered_df['Black'], label='Black')
ax.plot(filtered_df['Date'], filtered_df['White'], label='White')
ax.plot(filtered_df['Date'], filtered_df['Hispanic'], label='Hispanic')

ax.legend()

# Add horizontal line
ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='grey')

# Set y-axis values to percentages
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

# Remove borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

fig = plt.gcf()
fig.text(
    0.10, 0.93, 'Vaccination Rates (%) of Different Racial/Ethnic Groups in Harris County During 2021',
    fontname='Georgia', fontsize=13, weight='bold'
)


```
